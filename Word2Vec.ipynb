```python
from gensim.models import Word2Vec
```

```python
with open("text8", "r", encoding="utf-8") as f:
    text = f.read()

# Convert to list of words
words = text.split()

# Break into "sentences" of fixed length (important!)
sentences = [words[i:i+1000] for i in range(0, len(words), 1000)]

print("Number of sentences:", len(sentences))
```

```python
model = Word2Vec(
    sentences=sentences,
    vector_size=200,
    window=5,
    min_count=5,
    workers=4,
    sg=1,        # 1 = Skip-gram (recommended for text8)
    hs=1,        # Hierarchical Softmax
    epochs=5
)
```

```python
model.wv.most_similar(
    positive=["king", "woman"],
    negative=["man"],
    topn=5
)
```

```python
model.save("word2vec_text8.model")
```
